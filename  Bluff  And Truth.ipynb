```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv('Position_Salaries.csv')
X=dataset.iloc[:,1:2].values
y=dataset.iloc[:,2]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor=LinearRegression()
regressor.fit(X,y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_reg=PolynomialFeatures(degree=4)
X_poly=poly_reg.fit_transform(X)
poly_reg.fit(X_poly,y)
lin_reg2=LinearRegression()
lin_reg2.fit(X_poly,y)
```

```python
plt.scatter(X,y,color='red')
plt.plot(X,regressor.predict(X),color='blue')
plt.title('bluff & Truth (Multiple regression)')
plt.xlabel('Position level')
plt.ylabel('salary')
plt.show()
```

```python
plt.scatter(X,y,color='red')
plt.plot(X,lin_reg2.predict(X_poly),color='blue')
plt.title('bluff & Truth (Polynomial regression)')
plt.xlabel('Position level')
plt.ylabel('salary')
plt.show()
```

```python
#the Accuracy
print('The accuracy of multiple Linear regression=',regressor.score(X,y))
print('The accuracy of polynomial  Linear regression=',lin_reg2.score(X_poly,y))
```
